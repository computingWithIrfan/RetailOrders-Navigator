#### Importing libraries

```python
import pandas as pd
import numpy as np
```

#### Reading fetched csv file and creating it's dataframe

```python
orders_data = pd.read_csv('./dataset/orders.csv')
orders_data.head(10)
```

#### Hanlding Missing data and Renaming columns

##### - We can see that there two values **Not Available** and **unknown** in Ship Mode column which should also be treated as **Nan**

```python
orders_data['Ship Mode'].unique()
```

##### - **Solution** : We will pass na_value arguement at the time of reading_csv.

```python
orders_data = pd.read_csv('./dataset/orders.csv',na_values=['Not Available','unknown'])
orders_data.head(10)
```

##### - We should also refactor column names (To lowercase and separated by underscore if contain multiple words)

```python
orders_data.columns = orders_data.columns.str.lower()
orders_data.columns = orders_data.columns.str.replace(' ','_')
orders_data.head(3)
```

 #### Deriving new columns (discount, sale_price, profit)

```python
orders_data['discount'] = orders_data['list_price']* orders_data['discount_percent'] * 0.01
orders_data['sale_price'] = orders_data['list_price'] - orders_data['discount']
orders_data['profit'] = orders_data['sale_price'] - orders_data['cost_price']
orders_data.head()
```

#### Converting order_date data type from obj to datetime

```python
orders_data['order_date'] = pd.to_datetime(orders_data['order_date'], format='%Y-%m-%d')
orders_data.dtypes
```

#### Droping unnecessary columns (cost_price ,list_price and discout_percent)

```python
orders_data.drop(columns=['discount_percent','cost_price','list_price'],inplace =  True)
orders_data.head()
```
